```python
import rasterio as rio
from rasterio.plot import show
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
```

```python
# Reading the Full Oil_palm image to 16bit
```

```python
with rio.open("data/16bit/oil-palm-clip.tif") as src:
    data = src.read()
    profile = src.profile
```

```python
profile
```

```python
profile['dtype'] = 'uint8'
```

```python
data_new = ((data/data.max())*255).astype('uint8').transpose(1,2,0) 
```

```python
plt.imshow(data_new)
plt.show()
```

```python
# Saving the image as tiff
```

```python
with rio.open("data/8bit/oil_palm_8bit.tif", 'w', **profile) as dst:
    dst.write(data_new.transpose(2, 0, 1))
```
