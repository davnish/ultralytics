```python
import rasterio as rio
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.patches import Rectangle
import geopandas as gpd
```

```python
with rio.open("preprocessing_data/8bit/patched/8bit.3.tif") as src:
    data = src.read()
```

```python
df = pd.read_csv("preprocessing_data/inter/label_txt/8bit.3.txt", delimiter = " ", names = ['class', 'x', 'y', 'width', 'height'])
df
```

```python
data = data.transpose(1,2,0)
data.shape
```

```python
plt.imshow(data)
plt.show()
```

```python
bbox = df.iloc[:, 1:3]
```

```python
bbox.head(5) 
```

```python
ax = plt.gca()
ax.imshow(data)
for x, y in bbox.values:
    ax.add_patch(Rectangle(((x*256) - 8, (y*256) - 8), 16, 16, linewidth=1, edgecolor='blue', facecolor='none'))
#     print(x, y)
plt.show()
```
